```python
import json
from pymongo import MongoClient

client = MongoClient('localhost',27017) 
```

```python
#create database and collections
db = client['Final_Project']
GME = db['GME']
AMC = db['AMC']

with open('GME1.json') as g:
    file_data1 = json.load(g)

with open('AMC1.json') as g:
    file_data2 = json.load(g)
```

```python
#insert documents to collections
GME.insert_many(file_data1)
AMC.insert_many(file_data2)
```

```python
#Check the documents in collections
GME_list = GME.find()
for i in GME_list:
    print(i)
    
AMC_list = AMC.find()
for i in AMC_list:
    print(i)
```

```python
#Change data type
query1 = GME.aggregate([ 
    {"$set":{"date":{"$dateFromString":{"dateString":"$date"}}}},
    {"$out" : "GME"}])

```

```python
#Change data type
query2 = AMC.aggregate([ 
    {"$set":{"date":{"$dateFromString":{"dateString":"$date"}}}},
    {"$out" : "AMC"}])
```

```python
# price changes for a given ticker in 2020
import datetime
# GME
myquery1 = {'date': {'$gte': datetime.datetime(2020,1,1), 
                     '$lt': datetime.datetime(2021,1,1)}}

doc1 = GME.find(myquery1)
for x in doc1:
    print(x['date'].strftime("%Y-%m-%d"),x['close'])
```

```python
# top n most active dates
#GME
query3 = db.GME.aggregate([
    {'$sort': {'volume': -1}},
    {'$limit': 5}
 ])

for x in query3:
    print(x['date'].strftime("%Y-%m-%d"),x['volume'])

# top n most active dates
# AMC
query3 = db.GME.aggregate([
    {'$sort': {'volume': -1}},
    {'$limit': 5}
 ])

for x in query3:
    print(x['date'].strftime("%Y-%m-%d"),x['volume'])
```
